```python
import os
import sys

# 1. Get the current working directory (CWD)
# This is usually the directory from which you started the Jupyter server or script
current_working_directory = os.getcwd()
print(f"Current Working Directory: {current_working_directory}")

# 2. Get the parent of that directory
# os.path.dirname returns the directory name portion of a path
parent_directory = os.path.dirname(current_working_directory)
parent_directory = os.path.dirname(parent_directory)
print(f"Parent Directory: {parent_directory}")

# 3. Set sys.path (insert the parent directory at the beginning of the system path)
# This allows Python to find modules located in the parent folder
sys.path.insert(0, parent_directory)

# Verification
print("\nSystem path updated:")
for i, path in enumerate(sys.path):
    print(f"[{i}]: {path}")
```
